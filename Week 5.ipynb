```python
df_fp_s=df[['Member gender', 'First party',]]
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

```python
gb_fp_s=df_fp_s.groupby(['First party', 'Member gender']).size()
```

```python
gb_fp_s
```

```python
gb_fp_s=gb_fp_s.reset_index()
gb_fp_s=gb_fp_s.rename(columns={0:'count'})
```

```python
gb_fp_s=gb_fp_s[gb_fp_s['count'] >= 6]
```
